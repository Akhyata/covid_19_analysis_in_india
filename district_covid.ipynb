```python
import pandas as pd
import requests
import json
```

```python
district_data_url = 'https://data.covid19india.org/v4/min/data.min.json'
district_data = requests.get(district_data_url)
data = district_data.json()
```

```python
district = []

# Loop over all states
for state_code, state_data in data.items():
    # Loop over all districts within each state
    for district_code, district_data in state_data.get('districts', {}).items():
        # Extract relevant data categories
        confirmed_total = district_data.get('total', {}).get('confirmed', 0)
        recovered_total = district_data.get('total', {}).get('recovered', 0)
        deceased_total = district_data.get('total', {}).get('deceased', 0)
        tested_total = district_data.get('total', {}).get('tested', 0)
        vaccinated1_total = district_data.get('total', {}).get('vaccinated1', 0)
        vaccinated2_total = district_data.get('total', {}).get('vaccinated2', 0)
         #delta
        confirmed_delta = district_data.get('delta', {}).get('confirmed', 0)
        recovered_delta = district_data.get('delta', {}).get('recovered', 0)
        deceased_delta = district_data.get('delta', {}).get('deceased', 0)
        vaccinated1_delta = district_data.get('delta', {}).get('vaccinated1', 0)
        vaccinated2_delta = district_data.get('delta', {}).get('vaccinated2', 0)
         #delta7
        confirmed_delta7 = district_data.get('delta7', {}).get('confirmed', 0)
        recovered_delta7 = district_data.get('delta7', {}).get('recovered', 0)
        deceased_delta7 = district_data.get('delta7', {}).get('deceased', 0)
        vaccinated1_delta7 = district_data.get('delta7', {}).get('vaccinated1', 0)
        vaccinated2_delta7 = district_data.get('delta7', {}).get('vaccinated2', 0)
        #meta
        population_meta = district_data.get('meta', {}).get('population', 0)
        tested_meta = district_data.get('meta', {}).get('tested', {}).get('date', 0)
        last_updated_meta = district_data.get('meta', {}).get('last_updated', 0)
        notes_meta = district_data.get('meta', {}).get('notes', 0)
        date_meta = district_data.get('meta', {}).get('date', 0)
        # Adding data to list of dictionaries
        district_dict = {
            'state_code': state_code,
            'district_code': district_code,
            #total
            'confirmed_total': confirmed_total,
            'recovered_total': recovered_total,
            'deceased_total': deceased_total,
            'tested_total': tested_total,
            'vaccinated1_total': vaccinated1_total,
            'vaccinated2_total': vaccinated2_total,
            #delta
            'confirmed_delta': confirmed_delta,
            'recovered_delta': recovered_delta,
            'deceased_delta': deceased_delta,
            'vaccinated1_delta': vaccinated1_delta,
            'vaccinated2_delta': vaccinated2_delta,
            #delta7
            'confirmed_delta7': confirmed_delta7,
            'recovered_delta7': recovered_delta7,
            'deceased_delta7': deceased_delta7,
            'vaccinated1_delta7': vaccinated1_delta7,
            'vaccinated2_delta7': vaccinated2_delta7,
            #meta
            'population_meta': population_meta,
            'tested_meta' : tested_meta,
            'last_updated_meta' : last_updated_meta,
            'notes_meta' : notes_meta,
            'date_meta' : date_meta

        }
        district.append(district_dict)

# Convert list of dictionaries to DataFrame
district = pd.DataFrame(district)

district.to_csv("district_covid_data.csv", index=False)
```
